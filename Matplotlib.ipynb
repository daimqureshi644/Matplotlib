**Bar Plot**

```python
from matplotlib import colors
import matplotlib.pyplot as plt

x = ["python","c","c++","java"]
y = [85,50,60,82]

plt.xlabel("language",fontsize = 15)
plt.ylabel("No:",fontsize = 15)
plt.title("Bar Graph",fontsize = 15)
c = ["y","b","m","g"]
plt.bar(x,y,width=0.3,color = c, edgecolor = "r",linewidth =2,linestyle = ":",alpha = 0.4)
plt.show()
```

```python
from matplotlib import colors
import matplotlib.pyplot as plt
import numpy as np

x = ["python","c","c++","java"]
y = [85,50,60,82]
z = [20,30,40,50]
width = 0.2
p = np.arange(len(x))
p1 =[j+width for j in p]

plt.xlabel("language",fontsize = 15)
plt.ylabel("No:",fontsize = 15)
plt.title("Bar Graph",fontsize = 15)

plt.bar(p,y,width,color = "r",label = "Qureshi")
plt.bar(p1,z,width,color = "y",label = "Qureshi1")

plt.xticks(p+width/2,x,rotation = 10)
plt.legend()
plt.show()
```

**Pie Plot/Chart**

```python
from matplotlib import colors
import matplotlib.pyplot as plt
import numpy as np

y = ["python","c","c++","java"]
x = [85,50,60,82]
ex = [0.0,0.2,0.0,0.0]
c = ["r","b","g","y"]


plt.pie(x,labels= y,explode=ex,colors= c,autopct="%0.1f%%",shadow=True,startangle = 230,textprops={"fontsize": 10},
        wedgeprops={"linewidth": 3, "edgecolor": "m"},rotatelabels= False)

#plt.xlabel("language",fontsize = 15)
#plt.ylabel("No:",fontsize = 15)
plt.title("pie plot",fontsize = 15)
plt.legend(loc = 2)

plt.show()
```

**Stem Plot**

```python
import matplotlib.pyplot as plt
import numpy as np

x = [85,50,60,82,65,98,77]
y = [10,20,30,40,50,60,70]

plt.stem(x,y,linefmt=":",markerfmt= "ro",basefmt="g",bottom= 0,label="python",orientation="horizontal")

plt.legend()
plt.show()
```

**Box plot vs wisker plot**

```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50,60,70]

plt.boxplot(x,labels=["python"],showmeans=True,sym="r*",boxprops=dict(color="g"),
            capprops=dict(color="r"),whiskerprops=dict(color="r"))


plt.show()
```

**Stackplot**

```python
import matplotlib.pyplot as plt
import numpy as np

x = [1,2,3,4,5]
y = [2,5,4,2,7]
y1 = [4,5,6,7,8]
y2 = [12,13,15,17,18]

l = ["y","y1","y2"]

plt.stackplot(x,y,y1,y2,labels=l)

plt.legend(loc=2)
plt.show()
```

**Stepplot**

```python
import matplotlib.pyplot as plt
import numpy as np

x = [1,2,3,4,5]
y = [2,5,4,2,7]


plt.step(x,y,color ="r",marker ="o",ms=10,mfc="g",label="python")

plt.xlabel("steps",fontsize = 15)
plt.ylabel("No:",fontsize = 15)
plt.title("step plot",fontsize = 15)
plt.grid()
plt.legend(loc = 2)

plt.show()
```

**Fill Between plot**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([2,5,4,2,7])


plt.plot(x,y,color ="r")
plt.fill_between(x,y,color= "r", where=(x>=2)&(x<=4),alpha = 0.5)

plt.xlabel("x_axis",fontsize = 15)
plt.ylabel("y_asis",fontsize = 15)
plt.title("plot",fontsize = 15)


plt.show()
```
